```
# Task 1: Number Guessing Game
import random

def number_guessing_game():
    number = random.randint(1, 10)
    guess = None
    while guess != number:
        guess = int(input("Guess a number between 1 and 10: "))
        if guess < number:
            print("Too low.")
        elif guess > number:
            print("Too high.")
    print("Correct!")

# Run the game
number_guessing_game()
```

```
# Task 2: Count Vowels and Consonants
def count_vowels_and_consonants(s):
    vowels = 'aeiouAEIOU'
    v_count = c_count = 0
    for char in s:
        if char.isalpha():
            if char in vowels:
                v_count += 1
            else:
                c_count += 1
    print("Vowels:", v_count, "Consonants:", c_count)

# Input and function call
user_input = input("Enter a string: ")
count_vowels_and_consonants(user_input)
```

```
# Task 3: Music Library Management

def add_album(library, album_info):
    artist, title, year, songs = album_info
    library[title] = (artist, year, songs)
    return library

def create_playlist(library, song_selections):
    return [song for album, song in song_selections if song in library.get(album, (None, None, []))[2]]

def add_song_to_album(library, album_title, new_song):
    artist, year, songs = library[album_title]
    if new_song not in songs:
        songs.append(new_song)
    library[album_title] = (artist, year, songs)
    return library

def remove_song_from_playlist(playlist, song):
    if song in playlist:
        playlist.remove(song)
    return playlist

# Example usage
library = {}
album_info = ("Pink Floyd", "The Dark Side of the Moon", 1973, ["Speak to Me", "Breathe", "On the Run"])
library = add_album(library, album_info)
print("Library after adding album:", library)

song_selections = [("The Dark Side of the Moon", "Breathe")]
playlist = create_playlist(library, song_selections)
print("Playlist created:", playlist)

library = add_song_to_album(library, "The Dark Side of the Moon", "Time")
print("Library after adding song:", library)

playlist = remove_song_from_playlist(playlist, "Time")
print("Playlist after removing song:", playlist)
```

```
# Task 4: Simulate String Utility Package
def reverse_string(s):
    return s[::-1]

def to_uppercase(s):
    return s.upper()

def string_length(s):
    return len(s)

def is_palindrome(s):
    return s == s[::-1]

def is_alpha(s):
    return s.isalpha()

# Input and function calls
s = input("Enter a string: ")
print("Reversed:", reverse_string(s))
print("Uppercase:", to_uppercase(s))
print("Length:", string_length(s))
print("Is palindrome?", is_palindrome(s))
print("Is alphabetic?", is_alpha(s))
```

```
# Task 5: Convert string to int with try-except
def convert_to_int(s):
    try:
        return int(s)
    except ValueError:
        print("Invalid input.")
        return None

s = input("Enter a number: ")
result = convert_to_int(s)
if result is not None:
    print("Converted integer:", result)
```

```
# Task 6: Keep asking for valid int
def get_valid_integer():
    while True:
        try:
            return int(input("Enter an integer: "))
        except ValueError:
            print("Invalid input. Try again.")

num = get_valid_integer()
print("You entered:", num)
```

```
# Task 7: Bank Account Class
class BankAccount:
    def __init__(self, acc_no, name, balance=0):
        self.__acc_no = acc_no
        self.__name = name
        self.__balance = balance

    def get_info(self):
        return self.__acc_no, self.__name, self.__balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient balance")

# Example usage
acc = BankAccount("123", "John Doe", 1000)
acc.deposit(500)
acc.withdraw(300)
print("Account info:", acc.get_info())
```

```
# Task 8: Employee Class
class Employee:
    def __init__(self, emp_id, name, salary):
        self.__emp_id = emp_id
        self.__name = name
        self.__salary = salary

    def get_info(self):
        return self.__emp_id, self.__name, self.__salary

    def apply_hike(self, percentage):
        self.__salary += self.__salary * (percentage / 100)

# Example usage
emp = Employee("E101", "Alice", 50000)
emp.apply_hike(10)
print("Employee info after hike:", emp.get_info())
```
